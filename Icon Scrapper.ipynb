```python
from bs4 import BeautifulSoup
import requests
import os
import shutil
import glob
from PIL import Image
```

```python
urls = {'ability': ['https://dota2.fandom.com/wiki/Category:Ability_icons', 'https://dota2.fandom.com/wiki/Category:Ability_icons?filefrom=Fate%27s+Edict+icon.png#mw-category-media', 'https://dota2.fandom.com/wiki/Category:Ability_icons?filefrom=Phantasm+icon.png#mw-category-media', 'https://dota2.fandom.com/wiki/Category:Ability_icons?filefrom=Time+Lapse+icon.png#mw-category-media'],
        'custom_ability': ['https://dota2.fandom.com/wiki/Category:Custom_ability_icons', 'https://dota2.fandom.com/wiki/Category:Custom_ability_icons?filefrom=Eldwurm%27s+Touch+Marksmanship+icon.png#mw-category-media', 'https://dota2.fandom.com/wiki/Category:Custom_ability_icons?filefrom=Pale+Augur+Decay+icon.png#mw-category-media'],
        'diretide': ['https://dota2.fandom.com/wiki/Category:Diretide_icons'],
        'frosthaven': ['https://dota2.fandom.com/wiki/Category:Frosthaven_icons'],
        'old_ability': ['https://dota2.fandom.com/wiki/Category:Old_ability_icons'],
        'other_ability': ['https://dota2.fandom.com/wiki/Category:Other_ability_icons'],
        'unit_ability': ['https://dota2.fandom.com/wiki/Category:Unit_ability_icons']
       }
```

```python
if not os.path.exists('icons'):
    os.mkdir('icons')

for icon_type in urls.keys():
    if not os.path.exists(f'icons/{icon_type}'):
        os.mkdir(f'icons/{icon_type}')

if not os.path.exists('dataset/'):
    os.mkdir('dataset/')

if not os.path.exists('dataset/icons'):
    os.mkdir('dataset/icons')
```

```python
for icon_type in urls.keys():
    for url in urls[icon_type]:
        r = requests.get(url)
        soup = BeautifulSoup(r.text, 'html.parser')
        images = soup.findAll('img')
        
        for image in images:
            if image['src'].startswith('https://static.wikia.nocookie.net/'):
                print(image['src'].replace('/scale-to-width-down/120','/scale-to-width-down/128'))
                icon = requests.get(image['src'].replace('/scale-to-width-down/120','/scale-to-width-down/128')).content
                with open(f"icons/{icon_type}/{image['alt']}", "wb+") as f:
                                f.write(icon)
```

```python
# Deleting non-standard icons
for icon_path in glob.iglob("icons/*/*"):
    with Image.open(icon_path) as icon:
        if icon.size != (128, 128):
            print(icon_path, icon.size)
            os.remove(icon_path)
```

```python
# Creating a class folder for data loader to use with selected icons

for icon_type in ['ability', 'custom_ability', 'old_ability', 'other_ability', 'unit_ability']:
    source_dir = f'icons/{icon_type}'
    dst = 'dataset/icons/'
    files = glob.iglob(os.path.join(source_dir, "*.png"))
    for file in files:
        if os.path.isfile(file):
            shutil.copy2(file, dst)
```
